```python
import mouse
import keyboard
import time
import random

from pynput.mouse import Button, Controller
mouse2 = Controller()


total_questions = 827
period_questions = 16
problem_questions = 7

current_question = 821

while current_question <= total_questions:
    
    # refresh
    mouse.move(100,60)
    mouse.click('left')
    time.sleep(5)

    # expert
    mouse.move(150,250)
    mouse.click('left')
    time.sleep(0.3)
    mouse2.scroll(0, -25)
    time.sleep(1)
    mouse.move(190,830)
    mouse2.press(Button.left)
    mouse.move(174,830)
    mouse2.release(Button.left)
    mouse.move(150,830)
    mouse2.scroll(0, 40)
    time.sleep(1)

    # models
    mouse.move(250,250)
    mouse.click('left')
    time.sleep(0.3)
    mouse.move(190,490)
    mouse.click('left')
    keyboard.press_and_release('ctrl+a')
    keyboard.write('25')

    # chat
    mouse.move(150,220)
    mouse.click('left')
    time.sleep(0.3)


    for j in range(period_questions):

        for i in range(problem_questions):

            # activate excel
            mouse.move(1893,38)
            mouse.click('left')

            # copy input
            keyboard.press_and_release('down')
            time.sleep(1)
            keyboard.press_and_release('ctrl+c')
            time.sleep(1)

            # activate browser
            mouse.move(200,300)
            mouse.click('left')
            time.sleep(1)

            # paste input
            keyboard.press_and_release('ctrl+v')
            time.sleep(1)

            # run chat
            keyboard.press_and_release('enter')
            pause = random.randint(13, 17)
            time.sleep(pause)

            current_question = current_question + 1

            if current_question > total_questions:
                break

        # save
        mouse.move(500,350)
        mouse.click('left')
        time.sleep(1)

        # clear
        mouse.move(600,350)
        mouse.click('left')
        
        if current_question > total_questions:
            break

    # chat history
    mouse.move(550,220)
    mouse.click('left')
    time.sleep(0.3)
    mouse.move(550,360)
    mouse.click('left')
    time.sleep(1)
    mouse.move(350,450)
    mouse.click('left')
    time.sleep(1)    
```

```python
import json
import csv

input_filenames = ["6s2-232.json", "6s-435.json", "6s-554.json", "6s-a1.json", "6s-a2.json", "6s-a3.json", "6s-a4.json", "6s-a5.json", "6s-a6.json", "6s-a7.json", "6s-a8.json", "6s-a9.json", "6s-a10.json"]
input_filenames = ["86-106.json", "156-197.json", "289-330.json", "408-414.json", "674-694.json", "744-827.json", "863-869.json", "975-1030.json"]
input_filenames = ["a.json", "b.json"]
                   
output_filename = "falcon-results.csv"

data = []

for filename in input_filenames:
    # Opening JSON file
    f = open(filename)

    # returns JSON object as
    # a dictionary
    data.append(json.load(f))

    # Closing file
    f.close()

# Iterating through the json
# list

with open(output_filename, 'w', newline='', encoding='utf-8') as output_file:
    writer = csv.writer(output_file, delimiter=',', quotechar='"')

    for d in data:
        for key in d:
            questions = d[key][0]
            num = 1
            for question in questions:
                writer.writerow((question[0], question[1]))
                num = num + 1
        writer.writerow(("", ""))

output_file.close()
```
